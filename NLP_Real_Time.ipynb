```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk.corpus import stopwords
import re
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv('/content/hotel_review.csv')
```

```python
data.info()
```

```python
data.head()
```

```python
data['Review'][0]
```

# Lower case

```python
data['review_lower']= data['Review'].str.lower()
```

```python
data.head()
```

# Stop word Removal

```python
en_stopwords= stopwords.words('english')
print(en_stopwords)
```

```python
en_stopwords.remove('not')
```

```python
print(en_stopwords)
```

```python
import nltk
nltk.download('stopwords')
```

```python
data['review_no_stopwords']= data ['review_lower'].apply (lambda x: ' '. join([word for word in x. split() if word not in (en_stopwords)]))
```

```python
data.head()
```

```python
data['review_no_stopwords'][0]
```

# Punctuation

```python
data['review_no_stopwords_punc'] = data['review_no_stopwords'].apply(lambda x: re.sub(r'[*]','star', x))
```

```python
data.head()
```

```python
import re

data['review_no_stopwords_no_punc'] = data['review_no_stopwords_punc'].apply(lambda x: re.sub(r"[^\w\s]", "", str(x)))
```

```python
data.head()
```

```python
print(data.columns)


```

# Tokenizing

```python
data['tokenize']= data.apply(lambda x: word_tokenize(x['review_no_stopwords_no_punc']), axis = 1)
```

```python
data.head()
```

```python
data['tokenize'][0]
```

# Stemming

```python
ps = PorterStemmer()
```

```python
data['stemming']= data['tokenize'].apply(lambda tokens: [ps.stem(token) for token in tokens])
```

```python
data.head()
```

## Lemmatization

```python
lemmatizer = WordNetLemmatizer()
```

```python
data['lemmatized']= data['tokenize'].apply(lambda tokens: [lemmatizer.lemmatize(token) for token in tokens])
```

```python
data.head()
```

```python
import nltk
nltk.download('stopwords')
```

# N_Grams

```python
token_clean = sum (data['lemmatized'], [])
```

```python
print(token_clean)
```

```python
unigram=(pd.Series(nltk.ngrams(token_clean,1)).value_counts())
```

```python
print(unigram)
```

```python
bigram=(pd.Series(nltk.ngrams(token_clean,2)).value_counts())
```

```python
print(bigram)
```

```python
trigram=(pd.Series(nltk.ngrams(token_clean,3)).value_counts())
```

```python
print(trigram)
```
